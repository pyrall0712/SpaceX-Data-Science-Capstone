```python
from pyodide.http import pyfetch

async def download(url, filename):
    response = await pyfetch(url)
    if response.status == 200:
        with open(filename, "wb") as f:
            f.write(await response.bytes())

path = URL

await download(path, "dataset.csv")
file_name  = "dataset.csv"
```

```python
# Keep appending the code generated to this cell, or add more cells below this to execute in parts
import pandas as pd

# 1. Load Data
file_name = "dataset.csv"
df = pd.read_csv(file_name, header=0)

# 2. Menangani Data Kosong (Missing Values)
df['Screen_Size_cm'] = df['Screen_Size_cm'].fillna(df['Screen_Size_cm'].mode().iloc[0])
df['Weight_kg'] = df['Weight_kg'].fillna(df['Weight_kg'].mean())

# 3. Memperbaiki Tipe Data menjadi Float
df['Screen_Size_cm'] = df['Screen_Size_cm'].astype(float)
df['Weight_kg'] = df['Weight_kg'].astype(float)

# 4. Standardisasi: Konversi CM ke Inci dan KG ke Pound
df['Screen_Size_inch'] = df['Screen_Size_cm'] * 0.3937007874
df.drop(columns=['Screen_Size_cm'], inplace=True)

df['Weight_pounds'] = df['Weight_kg'] * 2.20462262185
df.drop(columns=['Weight_kg'], inplace=True)

# 5. Normalisasi CPU_frequency (Dividing by Max)
df['CPU_frequency'] = df['CPU_frequency'] / df['CPU_frequency'].max()

# 6. Mengubah Data Kategorikal (Screen) menjadi Numerik (Indicator Variables)
df = df.join(pd.get_dummies(df['Screen'], prefix='Screen')).drop(columns=['Screen'])

# 7. Practice 1: Konversi Harga USD ke Euro
df['Price'] = df['Price'] * 0.93

# 8. Practice 2: Normalisasi Min-Max pada CPU_frequency
# Catatan: Karena tadi sudah dibagi max, kita hitung ulang min-max nya
df['CPU_frequency'] = (df['CPU_frequency'] - df['CPU_frequency'].min()) / (df['CPU_frequency'].max() - df['CPU_frequency'].min())

# Menampilkan hasil akhir
print("Data Preparation Selesai!")
print(df.head())
```
